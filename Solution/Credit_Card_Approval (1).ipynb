Droping Co-related Columns in dataset

```python
data.drop(['CNT_CHILDREN','FLAG_WORK_PHONE','FLAG_PHONE','FLAG_EMAIL','ID'], axis=1, inplace=True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

y = data['target']
X = data.drop('target', axis=1)

x_train,x_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
import xgboost as xgb

#create XGboost Regressor
xgboost_model=xgb.XGBClassifier()

#train the model
xgboost_model.fit(x_train,y_train)


from sklearn.metrics import accuracy_score

#make Predictions
xgboost_predictions = xgboost_model.predict(x_test)
score=accuracy_score(y_test,xgboost_predictions)

print("score",score)
```

```python
from catboost import CatBoostClassifier

# Create a CatBoostClassifier object
model = CatBoostClassifier()

# Fit the model to the training data
model.fit(x_train, y_train)

# Predict the labels of the test data
y_pred = model.predict(x_test)

# Evaluate the model accuracy
accuracy_cbc = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy_cbc)
```

```python
from sklearn.ensemble import AdaBoostClassifier

# Create an AdaBoostClassifier object
abc = AdaBoostClassifier(n_estimators=50, learning_rate=1)

# Train the model
abc.fit(x_train, y_train)

# Make predictions
y_pred_abc = abc.predict(x_test)

# Evaluate the model
accuracy_abc = accuracy_score(y_test, y_pred_abc)
print("Accuracy:", accuracy_abc)

```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Create a GradientBoostingClassifier object
gbc = GradientBoostingClassifier()

# Train the model
gbc.fit(x_train, y_train)

# Make predictions
y_pred_gbc = gbc.predict(x_test)

# Evaluate the model
accuracy_gbc = accuracy_score(y_test, y_pred_gbc)
print("Accuracy:", accuracy_gbc)
```

```python
# K-Nearest Neighbors

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
# Create a KNN regressor
knn = KNeighborsClassifier(n_neighbors=int(np.sqrt(len(x_train))),metric='euclidean')

# Train the KNN regressor
knn.fit(x_train, y_train)

# Predict the life expectancy using the KNN regressor
y_pred_knn = knn.predict(x_test)

accuracy_knn=accuracy_score(y_pred_knn,y_test)

print("accuracy of knn",accuracy_knn)
```

```python
# logistic regression

from sklearn.linear_model import LogisticRegression

# Create a LogisticRegression object
logistic_regression = LogisticRegression()

# Train the model
logistic_regression.fit(x_train, y_train)

# Make predictions
y_pred_logistic = logistic_regression.predict(x_test)

# Evaluate the model
accuracy_logistic = accuracy_score(y_test, y_pred_logistic)
print("Accuracy:", accuracy_logistic)
```

```python
# table of accuracy of all algorithm and color which algorithm have high accuracy

import pandas as pd
import numpy as np

data = {'Algorithm': ['XGBoost', 'CatBoost', 'AdaBoost', 'GradientBoosting', 'KNN', 'Logistic Regression'],
       'Accuracy': [score, accuracy_cbc, accuracy_abc, accuracy_gbc, accuracy_knn, accuracy_logistic]}

df = pd.DataFrame(data)

# Sort the DataFrame by accuracy in descending order
df = df.sort_values(by='Accuracy', ascending=False)

# Set the index of the DataFrame to the algorithm names
df = df.set_index('Algorithm')

# Create a colormap for the table
cmap = plt.cm.coolwarm

# Color the table cells based on the accuracy values
df.style.background_gradient(cmap=cmap)
```
